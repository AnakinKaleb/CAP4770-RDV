First, I will begin by importing all modules that I think I will need and by reading the train.csv file.

```python
import pandas as pd
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.naive_bayes import GaussianNB
import os
from sklearn.model_selection import train_test_split,KFold
from sklearn.metrics import roc_auc_score, log_loss
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
train = pd.read_csv("train.csv", index_col = 0)
train
```

I begin by training 

```python
ctr=train["target"].value_counts()
ctr
```

Next I will display the data using a bar plot.

```python
import matplotlib.pyplot as plt
ctr.plot(kind="bar")
plt.show()
```

Next, I shape the data.

```python
train.shape
```

Next we fit the regression model to the training dataset.

```python
X=train.drop("target",axis=1)
y=train["target"]
le = LabelEncoder()
le_y = le.fit_transform(y)
logreg=LogisticRegression(multi_class='auto')
logreg.fit(X,le_y)
```

Next we will use the trained model to make predictions on the new test dataset.

```python
test = pd.read_csv("test.csv",  index_col = 0)
test
y_pred_prob=logreg.predict_proba(test)
pd_pre_prob=pd.DataFrame(y_pred_prob,columns=list(le.classes_))
pd_pre_prob
```

```python
submit = pd.read_csv("sampleSubmission.csv")
submit
```

Next we will fit a GaussianNB model to the dataset and use this model to make predictions on the test dataset.

```python
submission=pd.concat([submit['id'],pd_pre_prob],axis=1)
submission.to_csv("submission_1.csv",index=False)
gb= GaussianNB()
gb.fit(X,le_y)
y_pred_prob1=gb.predict_proba(test)
y_pred_prob1
```

Next we will use StratifiedKFold to perform kf-fold cross-validation on the model.

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=2022, )
gb= GaussianNB()    
results = cross_val_score(gb, X, le_y ,scoring='neg_log_loss', cv = kfold)
print(results)
print(results.mean())
```

Now we evaluate the performance of the model on the training dataset.

```python
pd_pre_prob1=pd.DataFrame(y_pred_prob1,columns=list(le.classes_))
pd_pre_prob1
```
